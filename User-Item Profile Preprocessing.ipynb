```python
# 특정 음식 뽑는 함수
def abstractFood(foodName, smallCate, df3):
    isspa = df3['식품대분류명'] == smallCate
    spa1 = df3[isspa]
    spa = spa1['식품명'].tolist()
    spagett = []
    for s in spa:
        if foodName in s:
            spagett.append(s)
    return spagett
```

```python
def select_value(val):    
    # (4) -> 피자, (8) -> 샐러드
    if '육류' in val:
        return random.choice(구이류)
    elif '치킨' in val:
        return random.choice(rest_chicken)
    elif '중화' in val:
        return random.choice(globals()[random.choice(rest_china)])
    elif '한식' in val:
        rest_k = random.choice(korea)
        rest_korea = globals()[rest_k]
        rest_korea = [*rest_korea]
        return random.choice(rest_korea)
    elif '회' in val:
        return random.choice(수조어육류)
    elif '이색' in val:
        # return random.choice(rest_other)
        return None
    elif '분식' in val:
        return random.choice(globals()[random.choice(rest_bunsik)])
    elif '채소' in val:
        return random.choice(salad)
    elif '피자' in val:
        return random.choice(onlyPizza)
    elif '빵' in val:
        return random.choice(dessert)
    # else:
    #     return 'none'
```

```python
def select_value_rest(val):
    if '음식점' in val:
        return random.choice(globals()[random.choice(korea)])
    elif '육류' in val:
        return random.choice(구이류)
    elif '치킨' in val:
        return random.choice(rest_chicken)
    elif '피자' in val:        
        return random.choice(pizza)
    elif '분식' in val:
        return random.choice(globals()[random.choice(rest_bunsik)])
    elif '일식' in val:
        return random.choice(globals()[random.choice(rest_japan)])
    elif '중식' in val:
        return random.choice(globals()[random.choice(rest_china)])
    elif '양식' in val:
        return random.choice(rest_)
    elif '이색' in val:
        # return random.choice(rest_other)
        return None
    elif '카페' in val:
        return random.choice(rest_cafe)
    elif '술' in val:
        return None
    # else:
    #     return 'none'
```

```python
def fill_random_value(df):
    for i in tqdm(range(df.shape[0])):
        tmp = df.iloc[i]
        # print(i, '번째')
        for j in range(6):
            name = select_value(tmp[j])
            print(name)
            if name != None:
                tmp[j] = foodName_swap[name]
        for j in range(6, 9):    
            name = select_value_rest(tmp[j])
            print(name)
            if name != None:
                tmp[j] = foodName_swap[name]
        
        # 중복 체크
        
        # 데이터프레임에 적용
        df.iloc[i] = tmp
       
```
